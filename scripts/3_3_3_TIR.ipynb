```python
import my_utils
import numpy as np
iteration = 1
dict_results = {}

def TIR(df):
    """
    Calculate the Time in Range (TIR) percentage for glucose levels.
    
    Parameters:
        df (pd.DataFrame): Input DataFrame containing glucose values.
        column (str): Name of the column containing glucose values.
    
    Returns:
        float: Percentage of glucose values within the target range (2.9 to 10 mmol/L).
    """
    # Filter rows where glucose is in the target range (2.9 < glucose <= 10)
    in_range = df[(df > 2.9) & (df <= 10)]
    
    # Calculate the percentage of in-range measurements
    percentage_in_range = (len(in_range) / len(df)) * 100
    
    return percentage_in_range


for (PtID, percentage), value in dictionary.items():
    

    # print('Iteration:',iteration,'PtID:',PtID,'ratio:' ,percentage)
    iteration=iteration+1
    print('patient ID: ',PtID, 'ratio: ', percentage)
    # My actual/true values and my baseline value
    y_actual = value['y_actual']
    y_last_value = value['y_last_val']
    y_pred_single = value['y_pred_single']
    y_pred_base = value['y_pred']
    y_pred_tl = value['y_pred_tl']
    y_last_value2 = y_last_value.values.reshape(-1, 1) # make y_last_val to nparray (672,1) from Series (672,)
    
    y_actual_tir = TIR(y_actual)
    y_last_value_tir = TIR(y_last_value)
    y_pred_single_tir = TIR(y_pred_single)
    y_pred_base_tir = TIR(y_pred_base)
    y_pred_tl_tir = TIR(y_pred_tl)
    # y_last_value2_tir = y_last_value.values.reshape(-1, 1) # make y_last_val to nparray (672,1) from Series (672,)

     
    dict_results[(PtID, percentage)] = {
        "ptid_test": value['PtId_test'],
        "ptid_group": value['ptid_group'],
        "y_actual": y_actual_tir,
        "y_last_val": y_last_value_tir,
        "y_pred_single": y_pred_single_tir,
        "y_pred_base": y_pred_base_tir,
        "y_pred_tl": y_pred_tl_tir
    
    }
```
