```python

def print_nested_dicts(d, indent=0):
    for key, value in d.items():
        if isinstance(value, dict):
            print('  ' * indent + f"{key}: ")
            print_nested_dicts(value, indent+1)
        else:
            print('  ' * indent + f"{key}: {value}")
            
            
print_nested_dicts(dictionary)
```
