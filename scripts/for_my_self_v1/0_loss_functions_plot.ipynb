```python

#%%
import pickle
import matplotlib.pyplot as plt

PtID = 179
ratio = 100
version = 3

file_path = f"/home/hbt/jchr_data/jchr_racial_diff/results/processed_data/2_1_1_predicted_results_rnn_wb/2_1_1_predicted_results_rnn_wb_v{version}/patient{PtID}_ratio{ratio}.pkl"
with open(file_path, 'rb') as file:
    dictionary = pickle.load(file)


nested_dict = dictionary.get((PtID, ratio))
epochs = nested_dict.get('epochs')
print(epochs)
print('goat')
```

```python


train_loss = nested_dict.get('loss')
val_loss = nested_dict.get('val_loss')
```

```python
plt.figure(figsize=(10, 5))

# Assuming the losses are stored in a list over epochs
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')

plt.title(f'Loss over epochs for PtID {PtID} at percentage {ratio}')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
```

```python
import matplotlib.dates as mdates
from datetime import datetime, timedelta

#%% plot CGM data
y_actual = nested_dict.get('y_actual')
y_pred = nested_dict.get('y_pred')
y_last_val = nested_dict.get('y_last_val')
y_pred_tl = nested_dict.get('y_pred_tl')


start_date = datetime.now()
num_days = 14

# Create a list of dates for each day in the range
dates = [start_date + timedelta(days=i) for i in range(num_days)]


# Plotting the data in subplots

plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)

plt.plot(y_actual,  marker='o', linestyle='-')
# plt.title('')
plt.xlabel('Year')
plt.ylabel('Revenue')
# plt.grid(True)

# Formatting the date to show the day number only
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d'))

# Set the locator to ensure each date is shown
plt.gca().xaxis.set_major_locator(mdates.DayLocator())


plt.xticks(rotation=45)


plt.subplot(2, 1, 2)
plt.plot(y_pred,  marker='o', linestyle='-')
# plt.title('')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.grid(True)

# Formatting the date to show the day number only
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d'))

# Set the locator to ensure each date is shown
plt.gca().xaxis.set_major_locator(mdates.DayLocator())

plt.xticks(rotation=45)




plt.show()
```
