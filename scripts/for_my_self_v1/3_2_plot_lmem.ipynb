```python
import statsmodels.formula.api as smf

df['race_W'] = df['Race'].map({'white': 'w', 'black': 'b'})
df['race_B'] = df['Race'].map({'black': 'b', 'white': 'w'})
df['age_C'] = df['AgeGroup'].map({1: 'c', 2: 'a'})
df['age_A'] = df['AgeGroup'].map({2: 'a', 1: 'c'})
df['sex_M'] = df['Gender'].map({'M': 'm', 'F': 'f'})
df['sex_F'] = df['Gender'].map({'F': 'f', 'M': 'm'})

df['ratio10'] = df['ratio'] / 10
df['train_size_base'] = (df['train_size_base'] /100000)-5.1
df

model = smf.mixedlm("rmse_base ~ ratio10 * race_W ", df, groups=df["PtID"])
# + sex_M + age_A + train_size_base

result = model.fit()

# Print the summary of the model
print(result.summary())

# # Compute confidence intervals for the coefficients
# conf_intervals = result.conf_int()
# conf_intervals.columns = ['2.5%', '97.5%']
# print(conf_intervals)

# # Combine the estimates, standard errors, and confidence intervals in one DataFrame
# coefficients = pd.DataFrame(result.params, columns=['Estimate'])
# coefficients['StdErr'] = result.bse
# coefficients['z value'] = result.tvalues
# coefficients['P value'] = result.pvalues
# coefficients[['2.5%', '97.5%']] = conf_intervals

# print(coefficients)
```
